```python

def get_results(start_index):
    headers = {
        'Accept': 'application/json',
        'user-key': '4acc54ca3cf4e3c53521751bd0a2ca40',
    }

    params = (
        ('entity_id', '305'),
        ('entity_type', 'city'),
        ('start', start_index ),
        ('count', 20)
    )
    
    url = 'https://developers.zomato.com/api/v2.1/search'
    response = requests.get(url, headers=headers, params=params)
    return response.json()


#x = get_results(0)

all_results = []

for x in range(100, 300, 20):
    try:
        # get the data
        temp_result = get_results(x)
        print(len(temp_result.get('restaurants')), "is the len of restaurants")
        all_results.append(temp_result)
        
    except Exception as e:
        print(e, x)
```
