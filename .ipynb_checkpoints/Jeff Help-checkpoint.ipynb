```python
def get_results(start_index):
    headers = {
        'Accept': 'application/json',
        'user-key': '18ac4342735c2d916c0c81f9cf6f407b',
    }

    params = (
        ('entity_id', '305'),
        ('entity_type', 'city'),
        ('start', start_index ),
        ('count', 20)
    )
    
    url = 'https://developers.zomato.com/api/v2.1/search'
    response = requests.get(url, headers=headers, params=params)
    return response.json()

#x = get_results(0)

all_results = []

for x in range(0, 200, 20):
    try:
        # get the data
        temp_result = get_results(x)
        print(len(temp_result.get('restaurants')), "is the len of restaurants")
        all_results.append(temp_result)
        
    except Exception as e:
        print(e, x)
```

```python
for_df = []

for row in all_results:
    for restaurant in row.get('restaurants'):
        for_df.append(restaurant.get('restaurant'))
    
df = pd.DataFrame.from_records(for_df)
df.tail()
```
